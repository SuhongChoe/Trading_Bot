```python

# OHLCV(open, high, low, close, volume)로 당일 시가, 고가, 저가, 종가, 거래량에 대한 데이터
# df = pyupbit.get_ohlcv("KRW-BTC", count=7)
df = fdr.DataReader('BTC/KRW', '2021-01-01')
print(df)

# 변동성 돌파 기준 범위 계산, (고가 - 저가) * k값
df['range'] = (df['High'] - df['Low']) * 0.5

# range 컬럼을 한칸씩 밑으로 내림(.shift(1))
df['target'] = df['Open'] + df['range'].shift(1)

# 수수료
fee = 0.0032
# np.where(조건문, 참일떄 값, 거짓일떄 값)
df['ror'] = np.where(df['High'] > df['target'],
                     df['Close'] / df['target'] - fee,
                     1)
# buy signal
df['buy_signal'] = np.where(df['High'] > df['target'],
                           df['Low']*0.97,
                           float("nan"))

# 누적 곱 계산(cumpord) => 누적 수익률
df['hpr'] = df['ror'].cumprod()

# Draw Down 계산 (누적 최대 값과 현재 hpr 차이 / 누적 최대값 * 100
df['dd'] = (df['hpr'].cummax() - df['hpr']) / df['hpr'].cummax() * 100

# 초기 자본금 100만원일 떄
df['money'] = df['hpr'] * 1000000

# MDD 계산
print("MDD(%): ", df['dd'].max())
# 엑셀로 출력
df.to_csv("dd.csv")
```

```python
plt.plot(df['money'])
plt.xticks(rotation = 45)
plt.show()
```

```python
# xlabels에 요일 추가
import datetime
_xticks = []
_xlabels = []
_wd_prev = 0
for _x, d in zip(x, btc.date.values):
    weekday = datetime.datetime.strptime(str(d), '%Y%m%d').weekday()
    if weekday <= _wd_prev:
        _xticks.append(_x)
        _xlabels.append(datetime.datetime.strptime(str(d), '%Y%m%d').strftime('%m/%d'))
    _wd_prev = weekday
axes[1].set_xticks(_xticks)
axes[1].set_xticklabels(_xlabels, rotation=45, minor=False)
```
